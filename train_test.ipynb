# load pretrained model

```python
import matplotlib.pyplot as plt
from alexnet import alexnet
import torch
model = alexnet()
optimizer = torch.optim.SGD(model.parameters(), 0.01,
                            momentum=0.9,
                            weight_decay=1e-4)

checkpoint = torch.load('model_best.pth.tar')
epoch = checkpoint['epoch']
best_acc1 = checkpoint['best_acc1']

print('Alexnet pretrained model.')
print('best accuracy(top 1), epoch {}: {:.2f}%'.format(epoch, best_acc1))

model.load_state_dict(checkpoint['state_dict'])
optimizer.load_state_dict(checkpoint['optimizer'])
```

```python
from PIL import Image
import numpy as np
cls_code = 'n02088364'
img_src = '/groups/pachitariu/pachitariulab/data/Share/Marius/ILSVRC2012/train/{}/{}_4904.JPEG'.format(cls_code, cls_code)
im = Image.open(img_src)
plt.imshow(np.asarray(im))
plt.axis('off')
plt.show()
```

```python
import torchvision.transforms as transforms

def image_loader(loader, image_name):
    image = Image.open(image_name).convert('L')
    image = loader(image).float()
    image = image.clone().detach().requires_grad_(True)
    image = image.unsqueeze(0)
    return image

normalize = transforms.Normalize(mean=[0.445],
                                 std=[0.269])

data_transforms = transforms.Compose([
            transforms.Grayscale(num_output_channels=1),
            transforms.RandomResizedCrop(224),
            transforms.RandomHorizontalFlip(),
            transforms.ToTensor(),
            normalize,
        ])

model.eval()
im = image_loader(data_transforms, img_src)
output = model(im)
pred = np.argmax(output.detach().numpy())
```
